```
 #imporing the module
import random
 
lst = ["Rock", "Paper", "Scissor"]
 
 
 
#displayng game rules
print("The Rules for the games are \n"
       + "Rock vs Paper -> Paper Wins \n"
       + "Paper vs Scissor -> Scissor Wins \n"
       + "Scissor vs Rock -> Rock Wins")
u_score = 0
c_score = 0
 
while True:
 
  print("Enter Your Choice \n 1.Rock \n 2.Paper \n 3.Scissor \n")
  choice = input()
  computer = lst[random.randint(0,2)]
 
  
  
  if choice == computer:
    print("your choice ", choice , " \n computer choice  " , computer)
    print("Tie\n")
    print("User Score is " , u_score ,"\n Computer Score is ", c_score)
  elif choice == "Rock":
    if computer == "Paper":
      print("your choice ", choice , " \n computer choice  " , computer)
      print("Computer Wins")
      c_score += 1
      print("User Score is " , u_score ,"\n Computer Score is ", c_score)
    else:
      print("your choice ", choice , " \n computer choice  " , computer)
      print("You Win")
      u_score += 1
      print("User Score is " , u_score ,"\n Computer Score is ", c_score)
  elif choice == "Paper":
    if computer == "Scissor":
      print("your choice ", choice , " \n computer choice  " , computer)
      print("Computer Wins")
      c_score += 1
      print("User Score is " , u_score ,"\n Computer Score is ", c_score)
    else:
      print("your choice ", choice , " \n computer choice  " , computer)
      print("You Win")
      u_score += 1
      print("User Score is " , u_score ,"\n Computer Score is ", c_score)
  elif choice == "Scissor":
    if computer == "Rock":
      print("your choice ", choice , " \n computer choice  " , computer)
      print("Computer Wins")
      c_score += 1
      print("User Score is " , u_score ,"\n Computer Score is ", c_score)
    else:
      print("your choice ", choice , " \n computer choice  " , computer)
      print("You Win")
      u_score += 1
      print("User Score is " , u_score ,"\n Computer Score is ", c_score)
  else:
    print("choose a valid choice")
 
  print("You Want to keep Playing \n Enter yes to keep playing  \n Enter no to stop playing")
  ans = input()
 
  if ans == 'no':
    break
print("Thanks For Playing \n BYE")
```
